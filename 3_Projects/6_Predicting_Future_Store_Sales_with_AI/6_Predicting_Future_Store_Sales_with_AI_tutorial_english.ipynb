### Theoretical Concept: What Is a Time Series?

A **time series** is a sequence of observations recorded in chronological order, such as daily sales, monthly website visits, or yearly revenue. Unlike many tabular datasets, the order of observations matters because each value may depend on previous values.

A time series is often described through four components:

1. **Trend:** The long-term upward or downward movement.
2. **Seasonality:** A repeating pattern over a fixed period, such as higher sales every December.
3. **Cycle:** Longer patterns that do not repeat on a fixed calendar schedule, such as economic cycles.
4. **Residuals:** Random variation left after trend and seasonality are removed.

### Step 3: Stationarity Testing

Before fitting ARIMA-style models, we need to check whether the series is **stationary**. A non-stationary series often needs transformation before modeling.

#### Theoretical Concept: Stationarity

A stationary time series has statistical properties that remain stable over time. In practical terms, its **mean**, **variance**, and **autocorrelation structure** do not drift.

Why this matters: ARIMA models are built around stable patterns in past values and forecast errors. If the mean keeps rising or the seasonal swings keep growing, the model can learn the wrong structure.

We check stationarity in two ways:

1. **Visual inspection:** Look for trend, seasonality, and changing variance.
2. **ADF test:** The Augmented Dickey-Fuller test uses a p-value to help decide whether the series is stationary.

For the ADF test:

- **Null hypothesis:** The series is non-stationary.
- **Alternative hypothesis:** The series is stationary.
- If `p-value < 0.05`, we usually reject the null hypothesis and treat the series as stationary.

#### Reading the ADF Test

The most important output is the **p-value**. A small p-value, usually below `0.05`, suggests the series is stationary. A large p-value means we do not have enough evidence to reject non-stationarity.

The test statistic and critical values provide another view: if the test statistic is more negative than a critical value, the result supports stationarity at that confidence level.

```python
def test_stationarity(timeseries):
    # Perform Dickey-Fuller test:
    print('Results of Dickey-Fuller Test:')
    dftest = adfuller(timeseries, autolag='AIC')
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
    for key,value in dftest[4].items():
        dfoutput['Critical Value (%s)'%key] = value
    print(dfoutput)

test_stationarity(df['Passengers'])
```

The line `adfuller(timeseries, autolag='AIC')` runs the ADF test. The `autolag='AIC'` option lets the test choose a suitable lag length using the Akaike Information Criterion.

**Result:** The p-value is very high, around `0.99`, so we fail to reject the null hypothesis. The original series is non-stationary and must be transformed before ARIMA modeling.

### Step 4: Making the Series Stationary

We use two common transformations:

1. **Log transformation:** Stabilizes variance when fluctuations grow with the level of the series.
2. **Differencing:** Subtracts the previous value from the current value to remove trend.

```python
# 1. Apply log transformation to stabilize the variance
df_log = np.log(df['Passengers'])

# 2. Apply differencing to remove the trend
df_diff = df_log.diff().dropna()

# Plot the stationary series
df_diff.plot()
plt.title('Stationary Time Series (Log-Differenced)')
plt.show()

# Retest for stationarity
test_stationarity(df_diff)
```

### ACF and PACF Intuition

- **ACF:** Measures correlation between the series and its lagged versions.
- **PACF:** Measures the direct correlation with each lag after accounting for intermediate lags.

These plots help us choose the `p` and `q` parameters for ARIMA.

For ARIMA:

- **`p`** is suggested by the PACF plot.
- **`d`** is the number of differencing steps.
- **`q`** is suggested by the ACF plot.

### Step 5: Model Identification with ACF and PACF Plots

#### Theoretical Concept: ACF and PACF

Autocorrelation plots summarize how current values relate to past values.

- **ACF** helps identify moving-average behavior and the `q` term.
- **PACF** helps identify autoregressive behavior and the `p` term.
- Since we difference the log series once, we start with `d = 1`.

```python
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 10))
plot_acf(df_diff, ax=ax1, lags=20)
plot_pacf(df_diff, ax=ax2, lags=20)
plt.show()
```

**Interpretation:** PACF cuts off sharply around lag 1, suggesting `p = 1`. ACF also weakens after lag 1, suggesting `q = 1`. A reasonable baseline is **ARIMA(1, 1, 1)**.

## Theoretical Concept: ARIMA

**ARIMA** stands for AutoRegressive Integrated Moving Average. It is written as **ARIMA(p, d, q)**:

- **p:** Number of autoregressive lag terms.
- **d:** Number of differencing steps used to make the series stationary.
- **q:** Number of moving-average error terms.

ARIMA is useful for trend-driven time series, but the basic form does not explicitly model seasonality.

### Step 6: Building the ARIMA Model

```python
# Split data into training and test sets
train_data = df_log[:'1958']
test_data = df_log['1959':]

# Build ARIMA model
model = ARIMA(train_data, order=(1, 1, 1), freq='MS')
arima_result = model.fit()

# Get forecast
forecast = arima_result.get_forecast(steps=len(test_data))
forecast_ci = forecast.conf_int()

# Plot the forecast
plt.figure(figsize=(14, 7))
plt.plot(df_log, label='Original Log Data')
plt.plot(forecast.predicted_mean, label='Forecast')
plt.fill_between(forecast_ci.index, forecast_ci.iloc[:, 0], forecast_ci.iloc[:, 1], color='k', alpha=.15)
plt.title('ARIMA Model Forecast')
plt.legend()
plt.show()
```

**ARIMA result:** The model follows the broad trend but misses the repeating seasonal peaks and troughs. This is expected because plain ARIMA does not include a seasonal component.

### Step 7: Building a SARIMA Model for Seasonality

#### Theoretical Concept: SARIMA

**SARIMA** extends ARIMA by adding seasonal terms. It is written as **SARIMA(p, d, q)(P, D, Q, m)**.

- `(p, d, q)` are the non-seasonal ARIMA parameters.
- `(P, D, Q)` are the seasonal autoregressive, differencing, and moving-average parameters.
- `m` is the length of one seasonal cycle. For monthly data with yearly seasonality, `m = 12`.

SARIMA is a strong choice when a series has both trend and repeating seasonal behavior.

```python
# Build SARIMA model
# We can find the optimal P, D, Q through a grid search, but common values are 1.
sarima_model = sm.tsa.statespace.SARIMAX(train_data,
                                          order=(1, 1, 1),
                                          seasonal_order=(1, 1, 1, 12),
                                          enforce_stationarity=False,
                                          enforce_invertibility=False,
                                          freq='MS') # Explicitly set frequency to suppress warnings
sarima_result = sarima_model.fit()

# Get forecast
sarima_forecast = sarima_result.get_forecast(steps=len(test_data))
sarima_forecast_ci = sarima_forecast.conf_int()

# Plot the forecast
plt.figure(figsize=(14, 7))
plt.plot(df_log, label='Original Log Data')
plt.plot(sarima_forecast.predicted_mean, label='SARIMA Forecast', color='red')
plt.fill_between(sarima_forecast_ci.index, sarima_forecast_ci.iloc[:, 0], sarima_forecast_ci.iloc[:, 1], color='r', alpha=.15)
plt.title('SARIMA Model Forecast')
plt.legend()
plt.show()
```

**SARIMA result:** This is a major improvement. The model captures both the upward trend and the yearly seasonal peaks and troughs.

### Step 8: Final Evaluation

```python
# Reverse the log transformation to get actual passenger numbers
original_test_data = np.exp(test_data)
sarima_predictions = np.exp(sarima_forecast.predicted_mean)

# Calculate RMSE
rmse = np.sqrt(mean_squared_error(original_test_data, sarima_predictions))
print(f"SARIMA Model RMSE: {rmse:.2f}")

# Plot final results
plt.figure(figsize=(14, 7))
plt.plot(df['Passengers'], label='Original Data')
plt.plot(sarima_predictions, label='SARIMA Forecast', color='red')
plt.title('Final Forecast vs. Actual Data')
plt.legend()
plt.show()
```

```python
# Calculate the rolling mean with a window size of 12
df['MovingAverage'] = df['Passengers'].rolling(window=12).mean()

# Plot the original data and the moving average
plt.figure(figsize=(14, 7))
plt.plot(df['Passengers'], label='Original Data', color='blue')
plt.plot(df['MovingAverage'], label='12-Month Moving Average', color='red', linestyle='--')
plt.title('Original Data vs. 12-Month Moving Average')
plt.xlabel('Year')
plt.ylabel('Number of Passengers')
plt.legend()
plt.show()
```

```python
from statsmodels.tsa.holtwinters import SimpleExpSmoothing, ExponentialSmoothing

# 2. Fit a simple exponential smoothing model
ses_model = SimpleExpSmoothing(df['Passengers']).fit(smoothing_level=0.2, optimized=False)
df['SimpleExpSmoothing'] = ses_model.fittedvalues

# 3. Fit a double exponential smoothing model
des_model = ExponentialSmoothing(df['Passengers'], trend='add', seasonal='add', seasonal_periods=12).fit()
df['DoubleExpSmoothing'] = des_model.fittedvalues

# 4. Plot the original data and the fitted values from both models
plt.figure(figsize=(14, 7))
plt.plot(df['Passengers'], label='Original Data', color='blue', alpha=0.7)
plt.plot(df['SimpleExpSmoothing'], label='Simple Exponential Smoothing', color='green', linestyle='--')
plt.plot(df['DoubleExpSmoothing'], label='Double Exponential Smoothing', color='purple', linestyle='-.')
plt.title('Original Data vs. Exponential Smoothing Models')
plt.xlabel('Year')
plt.ylabel('Number of Passengers')
plt.legend()
plt.show()
```

- **Moving average:** Smooths short-term noise and shows trend, but lags behind and does not forecast seasonality well.
- **Exponential smoothing:** Gives more weight to recent values. Simple smoothing is best for stable data, while trend-aware smoothing can follow a rising series.
- **ARIMA:** Models autocorrelation after differencing, but does not handle seasonality by default.
- **SARIMA:** Adds seasonal structure, making it better for data with repeating annual patterns.

#### Visual Comparison Summary

The moving average is useful for seeing the underlying trend, but it lags and misses seasonality. Double exponential smoothing reacts faster to trend but still struggles with repeated seasonal peaks. SARIMA performs best here because it models both trend and seasonality.

#### Why Stationarity Matters

Stationarity matters because many time series models assume that the future behaves like the past in a stable way. If the average level, variance, or autocorrelation changes over time, the model may learn a pattern that will not hold in the forecast period.

For this project, the original series is not stationary, so we use log transformation and differencing before fitting ARIMA. SARIMA then adds seasonal differencing and seasonal terms to better represent the annual cycle.
