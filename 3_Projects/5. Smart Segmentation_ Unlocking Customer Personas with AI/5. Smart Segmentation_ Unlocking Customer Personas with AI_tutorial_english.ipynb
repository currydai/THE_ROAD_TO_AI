### Theoretical Concept: What Is Unsupervised Learning?

In earlier projects, we mainly used **supervised learning**. Those datasets included a known target variable such as `SalePrice`, `Survived`, or a disease label. The model learned the relationship between input features and that known answer.

**Unsupervised learning** is different. The dataset does **not** provide a target label. Instead of predicting a known outcome, the goal is to discover hidden structure in the data.

**Clustering** is one of the most common unsupervised learning tasks. It groups data points so that customers inside the same cluster are similar, while customers in different clusters are meaningfully different.

### Theoretical Concept: How K-Means Works

K-Means is an iterative clustering algorithm. Imagine sorting a pile of mixed items into boxes:

1. **Choose `k`:** Decide how many groups you want.
2. **Initialize centroids:** Pick starting center points for the groups.
3. **Assign points:** Send each data point to the nearest centroid.
4. **Update centroids:** Recalculate each centroid as the average position of its assigned points.
5. **Repeat:** Continue assigning and updating until the centroids barely move.

When the algorithm stops, each customer belongs to one cluster, and each cluster has a centroid that represents its typical profile.

### Step 3: Segmentation Model 1 - Income and Spending Score

This is our primary segmentation model because the EDA showed a strong relationship between annual income and spending score.

## Theoretical Concept: The Elbow Method

How do we choose the best number of clusters, `k`, for K-Means? The **Elbow Method** is a practical starting point.

For each possible value of `k`, we calculate **WCSS**: Within-Cluster Sum of Squares. In scikit-learn, this is also called **inertia**. WCSS measures how far points are from their assigned cluster centroid. Lower WCSS means tighter clusters.

As `k` increases, WCSS always decreases. The key is to find the point where the improvement starts to slow down. This bend in the curve is the "elbow", and it often gives a good tradeoff between too few clusters and too many clusters.

### K-Means Initialization Options

The `init` parameter controls how starting centroids are selected:

- **`k-means++`:** The default and recommended option. It spreads initial centroids apart, usually producing more stable results.
- **`random`:** Randomly selects observations as initial centroids. It is simple but can produce weaker results.
- **Custom NumPy array:** Allows you to provide exact initial centroid coordinates when you have prior knowledge or need reproducibility.

```python
# 1. Select and scale the features
X1 = df[['Annual Income (k$)', 'Spending Score (1-100)']]
scaler1 = StandardScaler()
X1_scaled = scaler1.fit_transform(X1)

# 2. Implement the Elbow Method
wcss1 = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, init='k-means++', random_state=42, n_init=10)
    kmeans.fit(X1_scaled)
    wcss1.append(kmeans.inertia_)

# 3. Plot the Elbow Curve
plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), wcss1, marker='o', linestyle='--')
plt.title('Elbow Method for Income-Spending Segmentation')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('WCSS (Inertia)')
plt.xticks(range(1, 11))
plt.show()
```

### Interpreting WCSS

Think of each cluster as a group and the centroid as that group's center. WCSS adds up the squared distances from every member to its centroid.

A **low WCSS** means members are close to their group center, so the cluster is compact. A **high WCSS** means the group is more spread out.

K-Means tries to minimize WCSS. The Elbow Method uses this behavior to help us choose a useful value of `k`.

**Observation:** The elbow is clearly around **k = 5**. WCSS decreases sharply until five clusters, then the curve begins to flatten. This matches the visual structure we saw in the scatter plot.

```python
# Build and fit the final model
kmeans1 = KMeans(n_clusters=5, init='k-means++', random_state=42, n_init=10)
df['Income_Cluster'] = kmeans1.fit_predict(X1_scaled)
```

#### 3.1 Visualizing and Interpreting the Income-Based Segments

```python
plt.figure(figsize=(12, 8))
sns.scatterplot(data=df, x='Annual Income (k$)', y='Spending Score (1-100)',
                hue='Income_Cluster', palette='viridis', s=100, alpha=0.8, edgecolor='black')
plt.title('Customer Segments by Income and Spending')
plt.legend(title='Cluster')
plt.show()
```

```python
# Quantitative Persona Analysis
cluster_profiles1 = df.groupby('Income_Cluster')[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']].mean().round(2)
cluster_profiles1['Size'] = df['Income_Cluster'].value_counts()
print("--- Income-Based Cluster Profiles ---")
cluster_profiles1
```

#### 3.2 Income-Based Personas and Strategy

Using the quantitative profiles, we can define practical customer personas:

- **Cluster 0: Careful High Earners** - High income but low spending. They may respond to premium value bundles, guarantees, or loyalty benefits.
- **Cluster 1: Standard Customers** - Average income and average spending. This is the core customer base and is suitable for broad campaigns.
- **Cluster 2: Budget Shoppers** - Low income and low spending. They are likely price-sensitive and may respond to discounts.
- **Cluster 3: High-Value Targets** - High income and high spending. These customers are ideal for VIP programs and premium offers.
- **Cluster 4: Enthusiasts** - Lower income but high spending. They may be younger, trend-driven, and responsive to new product launches.

### Step 4: Segmentation Model 2 - Age and Spending Score

Next, we build a second segmentation model to reveal age-related customer behavior.

```python
# 1. Select and scale the features
X2 = df[['Age', 'Spending Score (1-100)']]
scaler2 = StandardScaler()
X2_scaled = scaler2.fit_transform(X2)

# 2. Implement the Elbow Method
wcss2 = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, init='k-means++', random_state=42, n_init=10)
    kmeans.fit(X2_scaled)
    wcss2.append(kmeans.inertia_)

# 3. Plot the Elbow Curve
plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), wcss2, marker='o', linestyle='--')
plt.title('Elbow Method for Age-Spending Segmentation')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('WCSS (Inertia)')
plt.xticks(range(1, 11))
plt.show()
```

**Observation:** The elbow is less sharp in this feature space, but **k = 4** is a reasonable and interpretable choice.

```python
# Build and fit the final model for age segmentation
kmeans2 = KMeans(n_clusters=4, init='k-means++', random_state=42, n_init=10)
df['Age_Cluster'] = kmeans2.fit_predict(X2_scaled)

# Visualize the new clusters
plt.figure(figsize=(12, 8))
sns.scatterplot(data=df, x='Age', y='Spending Score (1-100)',
                hue='Age_Cluster', palette='magma', s=100, alpha=0.8, edgecolor='black')
plt.title('Customer Segments by Age and Spending')
plt.legend(title='Cluster')
plt.show()
```

#### 4.1 Interpreting the Age-Based Segments

This model gives us a different customer lens:

- **Young High Spenders:** Customers under 40 with high spending scores.
- **Young Low-to-Mid Spenders:** Younger customers with lower or moderate spending.
- **Older Average Spenders:** Customers over 40 with moderate spending scores.
- **Older Low Spenders:** Older customers with very low spending scores.

### Step 5: Alternative Method - Hierarchical Clustering

To make the analysis richer, we briefly compare K-Means with another clustering technique.

#### Theoretical Concept: Hierarchical Clustering and Dendrograms

Hierarchical clustering builds a tree-like structure of clusters. Instead of choosing `k` first, we inspect a **dendrogram** to see how observations merge into larger groups.

To estimate the number of clusters, look for a tall vertical distance that is not crossed by horizontal merge lines. Drawing a horizontal cut through that region gives a possible cluster count.

```python
plt.figure(figsize=(20, 10))
dendrogram = sch.dendrogram(sch.linkage(X1_scaled, method='ward'))
plt.title('Dendrogram for Income-Spending Data')
plt.xlabel('Customers')
plt.ylabel('Euclidean Distances')
plt.axhline(y=6, color='r', linestyle='--') # Example cut-off line
plt.show()
```

**Insight:** The dendrogram also supports **5 clusters** for the income-spending data. A horizontal cut around height 6 crosses five major branches.
